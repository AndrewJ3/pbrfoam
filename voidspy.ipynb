# Functions

```python
def nut(ke,dissipation):
    cmu=0.09
    return cmu *((ke**2)/(dissipation))
def densitycalc(p,T):
    return 48.14*(p/T)*(1+(0.446*p/(T**(1.2))))**-1

def thermalcond(p,T):
    return 2.682e-3*(1+(1.123e-3*p))*(T**(0.71*(1-2e-4*p)))

def Rekta(density,u_mag,diameter_length,kinematic_visc,porosity):
    numer=(density*u_mag*diameter_length)
    denom=(kinematic_visc*(1-porosity))
    
    return numer/denom

def Nukta(Pr,Re,porosity):
    
    lside=1.27*(Pr**(1/3)/porosity)*Re**(0.36)
    rside=0.033*(Pr**(1/2)/porosity**(1.07))*Re**(0.86)
    
    return lside+rside

def Re_norm(density,u_mag,diameter_length,kinematic_viscosity):
    numer=(density*u_mag*diameter_length)
    denom=(kinematic_viscosity)
    return numer/denom

def Pe(Re,Pr):
    return Re*Pr

def dynamic_visc(T):
    return 3.647e-7 * (T**0.7)

def therm_diff(thermcond,cp,density):
    return thermcond/(cp*density)
    
def vol_sph(diameter):
    return (4/3)*np.pi*(diameter/2)**3

def surface_area_sph(diameter):
    return 4*np.pi*(diameter/2)**2

def surface_area_cyl(diameter):
    return np.pi*(diameter/2)**2

def turb_ke(u):
    I=.2
#     up=np.sqrt((1.0/3.0)*())
    U=np.sqrt(u**2)
    k=(3/2) *(I*U)**2
    return k
def turb_dissipation(turb_ke,l):
    cmu=0.09
    epst=((cmu**0.75)*(turb_ke**1.5))/(l)
    return epst
def hyd_dia(h,w):
    return 2*((h*w)/(h+w))
def DT_solid(ks,cps,rho_s):
    return ks/(cps*rho_s)
def Pr(nu,D_T):
    return nu/D_T
# def turb_length(hyd_dia):
#     return 
```

### Nusselt Number
### $Nu=1.27\frac{Pr^{1/3}}{\epsilon}Re^{0.36}+0.033\frac{Pr^{1/2}}{\epsilon^{1.07}}Re^{0.86}$
### Reynolds Number
### $Re=\frac{(\dot{m}/A)d}{\eta}$, $Re_p = Re/(1-\epsilon)$
### Prandtl Number
### $Pr=0.65975377969762$
### Peclet Number
### $Pe=RePr=UL/D_T$
### Thermal Conductivity
### $k=2.682\cdot10^{-3}(1+1.123 \cdot 10^{-3}p)\cdot T^{0.71(1-2\cdot10^{-4}p)} \ \ W/m \cdot K $
### Density
### $\rho=48.14\cdot \frac{p}{T}(1+0.4446\cdot \frac{p}{T^{1.2}})^{-1}$
### Dynamic Viscosity
### $\eta=3.674 \cdot 10^{-7} \cdot T^{0.7}$
### Kinematic Viscosity
### $\nu=\eta/\rho $
### Thermal Diffusivity
### $D_T=k/c_p \rho$
### Reynolds Number (kinematic viscosity)
### $Re= U L /\nu$
